```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook

d = pd.read_excel('origineel/week 5 meting 2 met zilver.xlsx', header=3)

a = pd.DataFrame()
a['I'] = d['I_plaat (A)']
a['U'] = d['U_H_versterkt'] / 527.5675676 * 10**6 #terug rekenen naar origineel in micro V
a['U'] = a['U'] - a['U'][0] # offset er af halen
```

```python
plt.plot(a['I'], a['U'], '.', label='data')

plt.legend()
plt.grid()
plt.xlabel('Stroom door plaat [A]')
plt.ylabel('Hall spanning U_H [mu V]')
```

```python
a.to_excel('output/meting2.xlsx', index=False)
```
